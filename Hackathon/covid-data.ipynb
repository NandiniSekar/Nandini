```python
import numpy as np
import pandas as pd
```

```python
covid = pd.read_csv(r'D:\nandu\COVID-19 Cases.csv')
covid.head(2)
```

```python
covid.shape
```

1. How many number of deaths in US

```python
len(covid[(covid['Case_Type']=='Deaths') & (covid['Country_Region']=='US') ])
```

2. To find the number of deaths province wise in US

```python
cod=covid[(covid['Case_Type']=='Deaths')& (covid['Country_Region']=='US')]
cod1=cod.loc[:,['Case_Type','Country_Region','Province_State']]
cod1.groupby(['Country_Region','Province_State']).agg({'Province_State':'count'})
```

3. To find the no. of confirmed cases in US

```python
cod=covid[(covid['Case_Type']=='Confirmed')& (covid['Country_Region']=='US')]
len(cod)
```

4. To find the number of confirmed cases province wise

```python
cod=covid[(covid['Case_Type']=='Confirmed')& (covid['Country_Region']=='US')]
cod1=cod.loc[:,['Case_Type','Country_Region','Province_State']]
cod1.groupby(['Country_Region','Province_State']).agg({'Province_State':'count'})
```
